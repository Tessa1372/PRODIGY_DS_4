```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
#loading the dataset

df = pd.read_csv(r'C:\Users\tessa\OneDrive\Documents\MDS\Datasets\twitter_sentiments.csv')
```

```python
# Exploring the dataset
print(df.head())
print(df.info())
print(df['Sentiment'].value_counts())

```

```python
#barplot
sns.countplot(x='Sentiment', data=df)
plt.title('Sentiment Distribution')
plt.show()
```

```python
#creating wordclouds
def plot_wordcloud(sentiment):
    plt.figure(figsize=(10, 6))
    # Converting 'Review' column to strings and replace NaN with an empty string
    text = ' '.join(df[df['Sentiment'] == sentiment]['Review'].astype(str).replace('nan', ''))
    wordcloud = WordCloud(width=800, height=400, random_state=21, max_font_size=110, background_color='white').generate(text)
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
    plt.title(f'Word Cloud for {sentiment} Sentiment')
    plt.show()

# Create a Counter for each sentiment
for sentiment in df['Sentiment'].unique():
    plot_wordcloud(sentiment)
```

```python
#plotting the statistics

from sklearn.feature_extraction.text import CountVectorizer
def plot_top_words(sentiment, n=10):
    text = ' '.join(df[df['Sentiment'] == sentiment]['Review'].astype(str).replace('nan', ''))
    vectorizer = CountVectorizer(stop_words='english')
    X = vectorizer.fit_transform([text])
    feature_names = vectorizer.get_feature_names_out()  
    word_freq = pd.DataFrame(list(zip(feature_names, X.toarray().flatten())), columns=['Word', 'Frequency'])
    top_words = word_freq.sort_values(by='Frequency', ascending=False).head(n)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Frequency', y='Word', data=top_words, palette='viridis')
    plt.title(f'Top {n} Words in {sentiment} Sentiment')
    plt.show()
for sentiment in df['Sentiment'].unique():
    plot_top_words(sentiment)
```

### Inference: As we can see from the above plots, words associated with 
### Negative sentiments include (hate, order etc) 
### Positive sentiments include (love,thank etc) 

##### Conclusion: Therefore using wordclouds and barplots we were able to analyze sentiments for a particular brand
